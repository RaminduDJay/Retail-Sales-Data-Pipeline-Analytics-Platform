```python
import pandas as pd
from sqlalchemy import create_engine
import seaborn as sns
import matplotlib.pyplot as plt
import os
from dotenv import load_dotenv

load_dotenv()

# PostgreSQL connection
engine = create_engine(f"postgresql://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}")
```

```python
# Load DBT model views
df_region = pd.read_sql("SELECT * FROM dbt_ramindu.sales_by_region", engine)
df_top_products = pd.read_sql("SELECT * FROM dbt_ramindu.top_products", engine)
```

```python
# Sales by Region
sns.barplot(x='region', y='total_sales', data=df_region)
plt.title("Total Sales by Region")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Top Products
sns.barplot(x='product_name', y='total_earned', data=df_top_products)
plt.title("Top Earning Products")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
